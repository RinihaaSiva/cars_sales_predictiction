# Merging the dataframes

```python
# pd.merge(df_1,df_2,on=cpmmon column_name,how=)
```

```python
emp=pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\Data Science\Learn Bay Classes\Practice Datasets\emp.csv")
dpt=pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\Data Science\Learn Bay Classes\Practice Datasets\dept.csv")
```

```python
emp
```

```python
dpt
```

```python
# how='inner' (default)
# It takes the common observation from both the dataframes
```

```python
pd.merge(emp,dpt,on='Depart_No',how='inner')
```

```python
# how='left'
# - ALl the observations from the left dataframe
```

```python
pd.merge(emp,dpt,on='Depart_No',how='left')
```

```python
# how='right'
# - All obsertvations from the right dataframe
```

```python
pd.merge(emp,dpt,on='Depart_No',how='right')
```

```python
# how='outer'
# - All observations from both the dataframes will be taken
```

```python
pd.merge(emp,dpt,on='Depart_No',how='outer')
```

```python
pd.merge(emp,dpt,on=['Depart_No','City'],how='inner')
```
